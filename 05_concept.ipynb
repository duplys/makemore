The likelihood function (often simply called the likelihood) is the probability density of observed data viewed as a function of the parameters of a statistical model.

Intuitively, the likelihood function $\mathcal{L}(\theta \mid x)$ is the probability of observing data  $x$ assuming  $\theta$ is the actual parameter. 

Consider a simple statistical model of a coin flip: a single parameter $p_\text{H}$ that expresses the "fairness" of the coin. The parameter is the probability that a coin lands heads up ("H") when tossed. $p_\text{H}$ can take on any value within the range 0.0 to 1.0. For a perfectly fair coin, $p_\text{H} = 0.5$.

Imagine flipping a fair coin twice, and observing two heads in two tosses ("HH"). Assuming that each successive coin flip is independent and identically distributed (i.i.d.), the probability of observing HH is:

$$P(\text{HH} \mid p_\text{H}=0.5) = 0.5^2 = 0.25$$

Equivalently, the likelihood of observing "HH" assuming $p_\text{H} = 0.5$ is:

$$\mathcal{L}(p_\text{H}=0.5 \mid \text{HH}) = 0.25$$


```python
import torch

L  = []
p_H = []

for i in range(0, 10):
    p = i/10
    p_H.append(p)
    L.append(p**2)

print(p_H)
print(L)
```

Now let's plot the likelihood function $\mathcal{L}$:

```python
import matplotlib.pyplot as plt

plt.plot(p_H, L)
plt.xlabel('p_H')
plt.ylabel('L')
plt.title('Likelihood function')
plt.show()
```

Next, let's look what log-likelihood looks like for $\mathcal{L}$:

```python
L2 = []

for e in L:
    L2.append(torch.log(torch.Tensor([e])))

plt.plot(p_H, L2)
plt.xlabel('p_H')
plt.ylabel('log-L')
plt.title('Log-likelihood function')
plt.show()
```

What about negative log-likelihood?

```python
L3 = []

for e in L:
    L3.append(-torch.log(torch.Tensor([e])))

plt.plot(p_H, L3)
plt.xlabel('p_H')
plt.ylabel('Negative log-L')
plt.title('Negative log-likelihood function')
plt.show()
```

Finally, let's look at what normalized negative log-likelihood looks like:

```python
L4 = []
n = 0

for e in L:
    L4.append(-torch.log(torch.Tensor([e])))
    n += 1

L4 = [e/n for e in L4]

plt.plot(p_H, L4)
plt.xlabel('p_H')
plt.ylabel('Normalized neg log-L')
plt.title('Normalized negative log-likelihood function')
plt.show()
```
